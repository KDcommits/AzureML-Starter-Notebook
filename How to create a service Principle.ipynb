<h1 align="center"> Service principle connection in Azure DevOps </h1>

References : 
- Follow the steps : https://www.youtube.com/watch?v=06wxMtSt_0g&ab_channel=sizeofeight"

**STEP 1 :**<br>
Go to Azure Active Directory -> Add -> App Registration -> Give the app name -> in Supported account types choosee Accounts in this organizational directory only (Default Directory only - Single tenant) -> Register.




**STEP 2 :**<br>
Go to `Certificates and Secret` and generate secret for the same. The secret key generation is important without which the connection to the azure devops won't be possible. 

**STEP 3:**<br>

In API permission I have selected `Grant admin consent for Deefault Directory` but I am not sure if it is at all required or not. So, before doing this just go through the documentation once, as I am not absolutely sure about the consequences.<br>



**Output  : The registered App**<br>

1. Object ID is the Service Princciple ID 

    
2. The value inside of secret and certificates is the Service Principle Key.<br>

    



**STEP 4:**<br>

Navigate to azure devops -> project Settings -> Service Connection -> Azure Resource Manager -> Service Principle(manual) -> Provide all the details that is necessary -> Verify -> Give a Service Connection Name -> Verify and Create.

**STEP 5 :**<br>

Now you have your service connection ready. <br>



**NOTE : A common Issue is that we fail to connect the service principle to Azure Devops and the verification fails**<br>

**Issue :** <br>
Failed to query service connection API: 'https://management.azure.com/subscriptions/da7f92e5-f637-4517-90fb-493d47170db2?api-version=2016-06-01'. Status Code: 'Forbidden', Response from server: '{"error":{"code":"AuthorizationFailed","message":"The client '9fd2d903-3411-4fe9-b466-dabe67f6df7a' with object id '9fd2d903-3411-4fe9-b466-dabe67f6df7a' does not have authorization to perform action 'Microsoft.Resources/subscriptions/read' over scope '/subscriptions/da7f92e5-f637-4517-90fb-493d47170db2' or the scope is invalid. If access was recently granted, please refresh your credentials."}}':<br>

**Resolution :** <br>
Assign 'Reader' Role to Service Principal: If the service principal does not have the 'Reader' role or appropriate permissions, you can assign it using the Azure CLI or Azure Portal.

Using Azure CLI: `az role assignment create --assignee-object-id 9fd2d903-3411-4fe9-b466-dabe67f6df7a --role "Reader" --scope "/subscriptions/da7f92e5-f637-4517-90fb-493d47170db2"`
